```python
import pandas as pd
from nltk import pos_tag
from nltk.probability import FreqDist
from nltk.corpus import treebank
import nltk
nltk.download("popular")
import string
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/orjinal.csv',index_col=0,dtype="string")
turkce_sutun_isimleri = ['yorum']
df.columns = turkce_sutun_isimleri
df.head()
```

```python
def kelime_birlestir(*edatlari_temizle):
    for a in edatlari_temizle:
        kelime=" ".join(a)
    return kelime
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize

TemizMetin=[]
pattern = r'[0-9]'
edatlar = stopwords.words('turkish')

for yorum in df.yorum:
  kucukharf=str(yorum).lower()
  sayikaldir = re.sub(pattern, '', kucukharf)
  noktalama = sayikaldir.translate(str.maketrans('','',string.punctuation))
  
  emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  
                           u"\U0001F300-\U0001F5FF"  
                           u"\U0001F680-\U0001F6FF"  
                           u"\U0001F1E0-\U0001F1FF"  
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u'\U00010000-\U0010ffff'
                           u"\u200d"
                           u"\u2640-\u2642"
                           u"\u2600-\u2B55"
                           u"\u23cf"
                           u"\u23e9"
                           u"\u231a"
                           u"\u3030"
                           u"\ufe0f"
                           "]+", flags=re.UNICODE)
  emoji_kaldir=emoji_pattern.sub(r'', noktalama)
  temiz_yorum=word_tokenize(emoji_kaldir)
  edatlari_temizle = [kelime for kelime in temiz_yorum if kelime not in edatlar]
  a=kelime_birlestir(edatlari_temizle)  
  TemizMetin.append(a)
yeni_df=pd.DataFrame(TemizMetin)  
yeni_df.to_csv('temiz_metin.csv',encoding="utf-8")

```
