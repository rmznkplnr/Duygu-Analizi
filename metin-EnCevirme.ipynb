```python
from textblob import TextBlob
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/yeni_temiz_Metin.csv',index_col=0,encoding="utf-8-sig")
turkce_sutun_isimleri = ['yorum']
df.columns = turkce_sutun_isimleri
df.head()
```

```python
en_data=[]
for y in df.yorum:
  text=TextBlob(str(y))
  print(text)
  text_en=text.translate(from_lang='tr',to="en")
  en_data.append(str(text_en))
yeni_df=pd.DataFrame(en_data)  
yeni_df.to_csv('turkce_metin_hazir.csv',encoding="utf-8-sig")
 


```
