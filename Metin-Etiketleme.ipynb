```python
from textblob import TextBlob
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/ingilizce_metin_merge.csv',index_col=0,encoding="utf-8-sig")
turkce_sutun_isimleri = ['yorum']
df.columns = turkce_sutun_isimleri
df.head()
```

```python
df.yorum
```

```python
# Create a function to get the subjectivity
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity(text):
  return TextBlob(text). sentiment. polarity

#Create two new columns
df['Subjectivity'] = df['yorum'].apply(getSubjectivity)
df['Polarity'] = df['yorum'].apply(getPolarity)

#Show the new dataframe with the new columns

df

```

```python
def getAnalysis(score):
  if score < 0:
    return 'Negative'
  elif score==0:
    return 'Neutral'
  else:
    return 'Positive'
df['Analysis'] = df['Polarity' ].apply(getAnalysis)
df
 
```

```python
df.to_csv("/content/drive/MyDrive/dataset/etkiketli.csv",encoding="utf-8-sig")
```
